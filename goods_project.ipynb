#  sales trends over time 

```python
df["Date"]=pd.to_datetime(df["Date"])
df
```

```python
df["months"]=df["Date"].dt.strftime("%B")
df["Year"]=df["Date"].dt.strftime("%Y")

df
```

```python
df["total_sale"]=df["Quantity Sold (liters/kg)"]*df["Price per Unit (sold)"]
df
```

# Total revenue generate

```python
sns.barplot(x="Brand",y="total_sale",data=df,palette="rocket",hue="Year")
plt.xticks(rotation=90)
plt.show()
```

# highest sale item

```python
sns.barplot(x="Product Name",y="total_sale",hue="Farm Size",data=df,palette="crest")
plt.xticks(rotation=99)
plt.show()
```

# profit mergan on each item

```python
profit_margen=df["total_sale"]- df["Price per Unit"]/100
profit_margen  
```

```python
sns.lineplot(x="Product Name",y="profit_margen",hue="Year",data=df,palette="cubehelix")
plt.xticks(rotation=99)
plt.show()
```

# Product Shelf Life Analysis: 

```python

df['Production Date'] = pd.to_datetime(df['Production Date'])
df['Expiration Date'] = pd.to_datetime(df['Expiration Date'])

df
```

```python




duration=(df["Expiration Date"]-pd.Timestamp.today()).dt.days
duration
```

```python

sns.barplot(x="Product Name",y="duration",hue="Farm Size",data=df,palette="crest")
plt.xticks(rotation=99)
plt.show()


```
